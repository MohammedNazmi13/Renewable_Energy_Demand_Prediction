### Theme: Energy | Domain: Prediction
Goal: Predict renewable energy demand using past data and weather conditions.


## Overview
This project uses meteorological factors like temperature, humidity, and solar radiation to forecast the demand for renewable energy. Supporting improved energy planning and maximizing the use of renewable resources are the objectives.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('/content/Renewable_Energy_Forecasting.csv')
data.head()
```

```python
data.info()
data.describe()
data.isnull().sum()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x='Temperature', y='Energy_Output_Class', data=data)
plt.title('Temperature vs Energy Output')
plt.show()

sns.heatmap(data.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

### Insights:
- As solar radiation and temperature increase, so does energy output.
- Energy demand and humidity have a weakly negative relationship.
- Consistent data is available for feature selection (Week 2).
